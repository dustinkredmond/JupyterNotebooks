### Scraping articles from Wikipedia using BeautifulSoup library

Install the *requests* and *bs4* (BeautifulSoup) libraries. And import them into our program.

```python
!!pip3 install requests bs4

import requests
from bs4 import BeautifulSoup
```

Use the below Wikipedia URL. This URL returns a random article on each visit.

```python
random_url = "https://en.wikipedia.org/wiki/Special:Random"
```

Loop five times, retrieving the title and URL of random Wikipedia articles.
Append these to our `results` list.

```python
results = []

for i in range(5):
    req = requests.get(random_url)
    soup = BeautifulSoup(req.content, "html.parser")
    
    #Title of Wikipedia Article found by "class" attribute
    title = soup.find(class_="firstHeading").text
    
    # Will now be populated with URL of the actual article
    url = req.url
    results.append({"title":title, "url":url})
   
```

Display the contents of our `results` variable. We now have a list of five random articles along with their title and URL.
